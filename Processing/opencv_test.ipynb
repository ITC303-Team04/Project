# GDAL: Retile
Place image files to be tiled in the input folder

```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.transform import rescale

window_size = '1500 1500'
input_dir  = './app/input/'
output_dir = './app/output/'
width = 1500
```

```python
images = os.listdir(input_dir)
rescaled_img = None
resized_img = None

for img_name in images:
    if img_name.startswith('Penrith'):
        continue
    try:

        img_path = os.path.join(input_dir, img_name)
        img = cv2.imread(img_path)
        print(img.shape)

        scale = width/img.shape[0]
        print(f"r e s c a l i n g")
        print(f"image width: {img.shape[0]}\nrescale width: {width}\nscale: {scale}\nimage shape: {img.shape}")
        rescaled_img = rescale(img.copy(), scale, anti_aliasing=False, channel_axis=True)

        print(f"r e s i z i n g")
        aspect = img.shape[1]/img.shape[0]
        resized_img = cv2.resize(img, (width,int(width/aspect)), interpolation=cv2.INTER_NEAREST)

        print(f"c o n v e r t i n g  c o l o r")
        resized_img = cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB)
        print(f"resize image shape: {resized_img.shape}")


    
    except Exception as e:
        print(f"e r r o r: {e}")


plt.imshow(resized_img)
plt.imshow(rescaled_img)
```
